```python
# Préparation : import des libraries et des données

import random
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Formation Python/Questions.csv',sep=';',encoding='latin1')

# Etape 1 : On crée une fonction qui pose les questions et renvoie le score

def askQuestion(quizData,score,nQuestion):
  print(quizData["Questions"][nQuestion])
  print("1- ",quizData["Réponse 1"][nQuestion])
  print("2- ",quizData["Réponse 2"][nQuestion])
  print("3- ",quizData["Réponse 3"][nQuestion])
  goodAnswer=quizData["Bonne réponse"][nQuestion]
  while True:
      try:
        userAnswer = int(input("Quelle est la bonne réponse ?"))
        if (userAnswer==goodAnswer) :
          print("Bravo, c'est la bonne réponse !")
          score += 1
        else :
          print("Mauvaise réponse, la bonne réponse était :",quizData.iloc[nQuestion,1+goodAnswer])
        break
      except ValueError:
        print("Oops! Ce n'est pas un nombre valide...")
  return score

# Etape 2 : On crée l'interface du jeu :
# - Introduction (explication des règles)
# - Choix des paramètres de la partie :
#   - Nombre de questions
#   - Score à atteindre pour gagner
# - Lancement des questions et gestion du score

# Introduction
print("Bienvenue dans ce quiz de culture générale !")
print("Chaque bonne réponse rapporte un point. Lorsque le score configuré est atteint tu as gagné")
print("Si tu n'atteins pas le score au bout du nombre de questions configurées tu as perdu")

# Choix des paramètres de la partie
while True:
      try:
        countQuestionMax = int(input("Combien de questions dans ce quiz ?"))
        scoreMax = int(input("Quel score pour gagner ?"))
        break
      except ValueError:
        print("Oops! Ce n'est pas un nombre valide...")

# C'est parti !
countQuestions = 0
score = 0
questionsList = list(data['Id'])
while countQuestions < countQuestionMax :
  if score == scoreMax :
    break
  countQuestions += 1
  print("")
  print("C'est parti pour la question n°",countQuestions)
  nQuestion = random.choice(questionsList)
  score = askQuestion(data,score,nQuestion)
  print("Score :",score)
  questionsList.remove(nQuestion)
if score == scoreMax :
  print("")
  print ("Félicitations tu as gagné en",countQuestions, "Questions !")
else :
  print("")
  print ("Désolé tu as perdu! Ton score final est de",score)
```
